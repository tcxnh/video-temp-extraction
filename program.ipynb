```python
import cv2
import os

# 输入视频文件路径
input_video = "/Users/yunzhu/Desktop/video temp extraction/input_video.mp4"

# 输出帧的目录
output_dir = "frames"
os.makedirs(output_dir, exist_ok=True)

# 打开视频文件
cap = cv2.VideoCapture(input_video)

# 获取视频的帧率
fps = int(cap.get(cv2.CAP_PROP_FPS))

# 设置每秒提取一帧
frame_interval = fps

frame_count = 0
saved_frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break  # 视频结束

    # 每秒提取一帧
    if frame_count % frame_interval == 0:
        frame_filename = os.path.join(output_dir, f"frame_{saved_frame_count:04d}.png")
        cv2.imwrite(frame_filename, frame)
        print(f"已保存帧: {frame_filename}")
        saved_frame_count += 1

    frame_count += 1

# 释放视频对象
cap.release()
print(f"帧已提取到目录: {output_dir}")
```

```python
import pytesseract
from PIL import Image
import re
import os
import cv2
import numpy as np
from collections import defaultdict
import csv

# 设置 Tesseract 路径（根据您的系统修改）
pytesseract.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'  # macOS/Linux
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Windows

# 定义提取最高温度和时间戳的函数
def extract_max_temperature_and_time(image_path):
    # 打开图像
    image = Image.open(image_path)
    width, height = image.size
    
    # 图像转换为OpenCV格式
    img_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    
    # 提取最高温度区域
    max_temp_region_cv = img_cv[50:100, 0:600]  # 只关注"最高"温度行
    
    # 提取时间戳区域
    time_region_cv = img_cv[height-100:height, 0:width]
    
    # 对最高温度区域进行处理
    max_temp_region_gray = cv2.cvtColor(max_temp_region_cv, cv2.COLOR_BGR2GRAY)
    # 二值化处理
    _, max_temp_region_bin = cv2.threshold(max_temp_region_gray, 127, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    max_temp_region_pil = Image.fromarray(max_temp_region_bin)
    
    # 对时间戳区域进行处理
    time_region_gray = cv2.cvtColor(time_region_cv, cv2.COLOR_BGR2GRAY)
    time_region_thresh = cv2.adaptiveThreshold(
        time_region_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    time_region_pil = Image.fromarray(time_region_thresh)
    
    # OCR提取最高温度
    max_temp_text = pytesseract.image_to_string(max_temp_region_pil, config='--psm 6 -l chi_sim+eng')
    max_temp_text_orig = pytesseract.image_to_string(
        Image.fromarray(max_temp_region_cv), config='--psm 6 -l chi_sim+eng')
    
    # OCR提取时间戳
    time_text = pytesseract.image_to_string(time_region_pil, config='--psm 6 -l eng')
    time_text_orig = pytesseract.image_to_string(
        Image.fromarray(time_region_cv), config='--psm 6 -l eng')
    
    # 打印调试信息
    print("最高温度区域OCR结果:", max_temp_text.strip())
    print("最高温度区域原始OCR结果:", max_temp_text_orig.strip())
    print("时间区域OCR结果:", time_text.strip())
    print("时间区域原始OCR结果:", time_text_orig.strip())
    print("=============")
    
    # 最高温度匹配模式
    temp_patterns = [
        r"最高[: 　]*(\d{1,2}\.\d)"        # 最高 28.6
        # r"(\d{1,2}\.\d)[ 　]*(?:℃|°C|°|C)",  # 28.6 ℃
        # r"(\d{1,2}\.\d)",                   # 直接匹配数字，如28.6
    ]
    
    max_temp = None
    for pattern in temp_patterns:
        # 在两种OCR结果中尝试匹配
        match = re.search(pattern, max_temp_text)
        if not match:
            match = re.search(pattern, max_temp_text_orig)
            
        if match:
            try:
                max_temp = float(match.group(1))
                print(f"匹配到最高温度: {max_temp}°C，使用模式: {pattern}")
                break
            except ValueError:
                continue
    
    # 时间戳匹配模式
    time_patterns = [
        r"(\d{2}):(\d{2}):(\d{2})",          # 00:00:31
        r"(\d{2})[\.:：](\d{2})[\.:：](\d{2})",  # 支持不同的分隔符
        r"(\d{2})(\d{2})(\d{2})",            # 直接连续数字000031
        r"(\d{4})[-/\.](\d{1,2})[-/\.](\d{1,2})[ 　]*(\d{1,2})[:\.:](\d{1,2})",  # 备用：2025/03/16 14:25格式
    ]
    
    timestamp = None
    for pattern in time_patterns:
        # 在两种OCR结果中尝试匹配
        match = re.search(pattern, time_text)
        if not match:
            match = re.search(pattern, time_text_orig)
            
        if match:
            # 根据匹配组合时间戳
            if len(match.groups()) == 3:  # 00:00:31 格式
                hour, minute, second = match.groups()
                timestamp = f"{hour.zfill(2)}:{minute.zfill(2)}:{second.zfill(2)}"
                print(f"匹配到时间戳: {timestamp}，使用模式: {pattern}")
                break
            elif len(match.groups()) == 5:  # 备用：2025/03/16 14:25 格式
                year, month, day, hour, minute = match.groups()
                timestamp = f"{year}/{month.zfill(2)}/{day.zfill(2)} {hour.zfill(2)}:{minute.zfill(2)}"
                print(f"匹配到时间戳: {timestamp}，使用模式: {pattern}")
                break
    
    # 如果上述区域没有匹配成功，尝试在整个图片中搜索
    if max_temp is None or timestamp is None:
        full_text = pytesseract.image_to_string(image, config='--psm 6 -l chi_sim+eng')
        print("完整图片OCR结果:", full_text.strip())
        
        if max_temp is None:
            for pattern in temp_patterns:
                match = re.search(pattern, full_text)
                if match:
                    try:
                        max_temp = float(match.group(1))
                        print(f"在完整图片中匹配到最高温度: {max_temp}°C")
                        break
                    except ValueError:
                        continue
        
        if timestamp is None:
            for pattern in time_patterns:
                match = re.search(pattern, full_text)
                if match:
                    if len(match.groups()) == 3:  # 00:00:31 格式
                        hour, minute, second = match.groups()
                        timestamp = f"{hour.zfill(2)}:{minute.zfill(2)}:{second.zfill(2)}"
                        print(f"在完整图片中匹配到时间戳: {timestamp}")
                        break
                    elif len(match.groups()) == 5:  # 备用：2025/03/16 14:25 格式
                        year, month, day, hour, minute = match.groups()
                        timestamp = f"{year}/{month.zfill(2)}/{day.zfill(2)} {hour.zfill(2)}:{minute.zfill(2)}"
                        print(f"在完整图片中匹配到时间戳: {timestamp}")
                        break
    
    # 可视化处理结果（调试用）
    cv2.imwrite("max_temp_region_debug.jpg", max_temp_region_bin)
    cv2.imwrite("time_region_debug.jpg", time_region_thresh)
    
    return max_temp, timestamp

# 主程序
def main():
    # 获取所有帧文件
    output_dir = "frames"  # 帧保存的目录
    
    # 确保目录存在
    if not os.path.exists(output_dir):
        print(f"目录 '{output_dir}' 不存在。创建目录...")
        os.makedirs(output_dir)
    
    frame_files = sorted([f for f in os.listdir(output_dir) if f.endswith((".png", ".jpg"))])
    
    if not frame_files:
        print(f"警告: 在 '{output_dir}' 目录中没有找到图片文件。")
        return
    
    # 存储最高温度和时间戳
    data = []
    
    for frame_file in frame_files:
        frame_path = os.path.join(output_dir, frame_file)
        print(f"\n处理帧: {frame_file}")
        
        max_temp, timestamp = extract_max_temperature_and_time(frame_path)
        
        if max_temp is not None and timestamp is not None:
            data.append({"time": timestamp, "max_temperature": max_temp})
            print(f"结果: 时间={timestamp}, 最高温度={max_temp}°C")
        else:
            print(f"警告: 未能从帧 {frame_file} 中提取最高温度或时间戳")
    
    if not data:
        print("错误: 未能从任何帧中提取数据。")
        return
    
    # 输出所有识别到的结果，不进行分组
    print("\n所有识别到的时间和最高温度对:")
    for entry in data:
        print(f"时间: {entry['time']}, 最高温度: {entry['max_temperature']}°C")
    
    # 保存所有结果到 CSV 文件
    output_csv = "all_temperatures.csv"
    with open(output_csv, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["时间", "最高温度 (°C)"])
        for entry in data:
            writer.writerow([entry["time"], entry["max_temperature"]])
    
    print(f"\n结果已保存到文件: {output_csv}")


main()
```
